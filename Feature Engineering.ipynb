### 2.2 Target Guided ordinal Categories 

```python
#Suppose for my Categories A o/p is 1, B is 1, C is 0, D is 1, again A is 0, B is 0
#It is repeated several number of times.
#What we do that, we are considering these f1 & o/p as a column.
#Each and every categories of feature f1, I'll tried out to find the mean of this particular values of 1 in o/p column.
#when I'm calculating the mean, It basically means, wherever the A value is 1 and wherever the A value is 0.
#I'm going to considered with respect to those features and I'm going to compute the mean of this value.
# when you see mean, You are trying to find out the number of values for which A is basically 1 and mean is giving for A is o.73, B is 0.6, C is 0.55
#Once you have the mean, what you have to do that, you have assign this rank on the basis of on particular mean.
#Here, I'm talking about oridinal categories, so we are assigning rank on particular mean.
#For A value most of the value is 1 and based on that A have higher mean.

#If A is having highest mean then we will provide highest rank.
#And on the basis of rank, we are providing label like 4,3,2,1,4,3,..... and this value is provided to ML algorithm.

#Because this ordinal, we are providing whole numbers, if it is Nominal then I cannot give this value directly because I don't know ordering in Nominal and I don't know which categorical features is more important.

#For that, I have another features that is MEAN ENCODING.
#If I have nominal Categories, How can we solve?






#Label	f1   O/P    Classification
#4	     A	  1  -> Mean 0.73
#3	     B	  1			 0.60
#2	     C	  0			 0.55
#1	     D	  1          0.33
#4	     A	  0          0.73
#3	     B	  0          0.60
```

## Feature Scaling

```python
## Why do we required Feature Scaling?
#Whenever we discussed about features scaling means, we discussed about features.
```

#### There are four common methods to perform Feature Scaling.

##### 1. Standardisation :- where mue(mean) = 0, Standard deviation = 1
##### 2. Mean Normalisation :- Values between -1 and +1, mue(mean) = 0
##### 3. Min-max Normalisation :- between 0 to 1
##### 4. Unit Vector :- range[0,1]


### Examples of Algorithms where Feature Scaling matters

##### 1. k-means uses the Euclidean distance measure here feature scaling matters.
##### 2. K-Nearest-Neighbours also require feature scaling.
##### 3. Principal Component Analysis(PCA): Tries to get the features with maximum variance, here too feature scaling is required.
##### 4. Gradient Descent (Like Linear Regression ): Calculation speed increase as Theta calculation becomes faster after feature scaling.
###### Note: Naive Bayes, Linear Discriminant Analysis, and Tree-Based Models are not affected by feature scaling. In short, any Algorithm which is not Distance based is not affected by Feature Scaling.
